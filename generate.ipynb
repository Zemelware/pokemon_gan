```python
import matplotlib.pyplot as plt
import numpy as np
from keras.models import load_model
```

```python
model = load_model('models/generator_final.h5')

# Keep the seed constant to generate the same image
# np.random.seed(100)

def latent_vectors(latent_dim, n_samples):
    latent_vec = np.random.randn(n_samples * latent_dim)
    latent_vec = latent_vec.reshape(n_samples, latent_dim)
    return latent_vec

r, c = 3, 5
noise = latent_vectors(128, r * c)
gen_imgs = model.predict(noise)

# Rescale images 0 - 1
gen_imgs = 0.5 * gen_imgs + 0.5

fig, axs = plt.subplots(r, c, figsize=(15, 8))
for i in range(r):
    for j in range(c):
        axs[i,j].imshow(gen_imgs[i*c + j])
        axs[i,j].axis('off')
```
